```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import math
```

```python
ratings_csv = pd.read_csv("dataset/ratings.csv")
books_csv = pd.read_csv("dataset/books.csv")
users_csv = pd.read_csv("dataset/users.csv")
```

```python
print('Number of users ' + str(len(users_csv['user_id'])))
print('Number of books ' + str(len(books_csv['book_id'])))
```

```python
bookInsights = pd.DataFrame(ratings_csv.groupby('book_id').size(), columns=['count'])
bookInsights.head()
```

```python
plt.plot(bookInsights.index, bookInsights['count'])
```

```python
popularity_thres = 1181
popular_books = list(set(bookInsights.query('count >= @popularity_thres').index))
ratings_csv_drop = ratings_csv[ratings_csv.book_id.isin(popular_books)]
print('shape of original ratings data: ', ratings_csv.shape)
print('shape of ratings data after dropping unpopular movies: ', ratings_csv_drop.shape)
print('Number of unique movies after dropping: ', len(ratings_csv_drop.book_id.unique()))
```

```python
ratings_csv_drop.head()
```

```python
print('Number of unique books: %d' %(len(ratings_csv_drop.book_id.unique())))
print('Number of unique users: %d' %(len(ratings_csv_drop.user_id.unique())))
```

```python
book_id_new = np.sort(ratings_csv_drop.book_id.unique())
books_csv_drop = books_csv[books_csv.book_id.isin(book_id_new)]
books_csv_drop.shape
```

```python
export_csv = ratings_csv_drop.to_csv(r'/Users/vishnusingh/Desktop/Sem 7/WT2/Project/ratings_drop.csv', index = None, header=True)
```
